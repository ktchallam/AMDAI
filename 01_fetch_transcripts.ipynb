# Module 2: Transcript Fetcher with Fallback Chain

This notebook demonstrates the `TranscriptFetcher` class, showing how it attempts to fetch transcripts for Reliance Industries and Tata Consultancy Services (TCS) using BSE, NSE, Trendlyne, or local Simulation fallbacks, and displays the source used and the word counts.

```python
# Cell 1: Import dependencies and test run
import sys
from loguru import logger

# Ensure logging output is visible
logger.remove()
logger.add(sys.stderr, level="INFO")

from transcript_fetcher import TranscriptFetcher

fetcher = TranscriptFetcher()
print("TranscriptFetcher initialized.")
```

```python
# Cell 2: Fetch Transcript for Reliance Industries
print("Fetching transcript for Reliance Industries...")
reliance_result = await fetcher.fetch(
    company_name="Reliance Industries",
    bse_code="500325",
    nse_symbol="RELIANCE",
    quarter="Q4FY26"
)

print("\n--- RELIANCE FETCH SUMMARY ---")
print(f"Company: {reliance_result.company}")
print(f"Quarter: {reliance_result.quarter}")
print(f"Source: {reliance_result.source}")
print(f"Fetched At: {reliance_result.fetched_at}")
print(f"Word Count: {len(reliance_result.text.split())}")
print(f"Snippet:\n{reliance_result.text[:500]}...")
```

```python
# Cell 3: Fetch Transcript for TCS
print("Fetching transcript for Tata Consultancy Services (TCS)...")
tcs_result = await fetcher.fetch(
    company_name="TCS",
    bse_code="532540",
    nse_symbol="TCS",
    quarter="Q4FY26"
)

print("\n--- TCS FETCH SUMMARY ---")
print(f"Company: {tcs_result.company}")
print(f"Quarter: {tcs_result.quarter}")
print(f"Source: {tcs_result.source}")
print(f"Fetched At: {tcs_result.fetched_at}")
print(f"Word Count: {len(tcs_result.text.split())}")
print(f"Snippet:\n{tcs_result.text[:500]}...")
```
